```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras 
import tensorflow_datasets as tfds
from keras.datasets import cifar10
print(tf.__version__)
```

```
#Dataset First CIFAR-10
(train_images,train_labels), (test_images, test_labels) = cifar10.load_data()
```

```
train_images = train_images /255.0
test_images = test_images /255.0
print(len(train_images))
print(len(test_images))
```

```
base_model_resnet50 = keras.applications.ResNet50V2(include_top= False, weights= None,input_shape=(32,32,3))
```

```
base_model_resnet50.summary()
```

```
x = base_model_resnet50.output
print(np.shape(x))
x = keras.layers.GlobalAveragePooling2D()(x)
print(np.shape(x))
x = keras.layers.Dropout(0.7) (x)
print("First ",np.shape(x))
x = keras.layers.Dense(units= 512, activation= keras.activations.relu) (x)
print("After 512 ",np.shape(x))
x = keras.layers.Dense(units= 128, activation= keras.activations.relu) (x)
print("After 128 ",np.shape(x))
x = keras.layers.Dense(units= 32, activation= keras.activations.relu) (x)
print("After 32 ",np.shape(x))
prediction_layer = keras.layers.Dense(units= 10, activation= keras.activations.softmax) (x)
print(np.shape(prediction_layer))
model = keras.Model(inputs = base_model_resnet50.input, outputs =  prediction_layer)
```

```
model.compile(loss="sparse_categorical_crossentropy", optimizer= keras.optimizers.Adam(),metrics=['accuracy'])
model.save('resnet50.h5')
```

```
#Time to train
num_epochs = 30

model.fit(x= train_images, y= train_labels, epochs= num_epochs, validation_data=(test_images, test_labels), batch_size= 64)
```
